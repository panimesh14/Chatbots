# Using OpenAI Python Package

```python
import openai #OpenAI Python Package
openai.api_key = "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"
messages = [
    #defines context for the api roles:system,user,assistant
    {'role':'system','content':'You are a Kind helpful assistant'} 
]
```

```python
#infinite loop for continuous unending chat 
while True:
    message = input("User : ")
    if message:
        messages.append({'role':'user','content':message}) #appending the role of user to the list of messages
        #response = openai.ChatCompletion.create(model="gpt-3.5-turbo", messages=messages)
        chat = openai.chat.completions.create(model="gpt-3.5-turbo",messages=messages)
    reply = chat.choices[0].message.content
    print(f"ChatGPT: {reply}")
    messages.append({"role":"assistant","content":reply})
```

Method Using API Endpoint

```python
import requests
URL = "https://api.openai.com/v/chat/completions"
payload = {
    "model":"gpt-3.5-turbo",
    "messages":[{"role":"user","content":f"Who is Bill Gates?"}],
    "temperature":1.0,
    "top_p":1.0,
    "n":1,
    "stream":False,
    "presence_penalty":0,
    "frequency_penalty":0
}
headers = {
    "Content-Type":"application/json",
    "Authorization":f"Bearer{openai.api_key}"
}
response = requests.post(URL, headers=headers, json=payload, stream=False)
```
